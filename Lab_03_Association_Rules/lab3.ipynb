```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules

df = pd.read_csv("Supermart.csv")
df.head()
#call libraries that are needed ands read the csv file and display it
```

```python
print(df.isnull().sum()) #Shows missing values. output shows that there are no missing values.
```

```python
# Make a new table called "products"
# Each row is a basket of Sub Category items

df["Order Date"] = pd.to_datetime(df["Order Date"], format="mixed")
df["OrderMonth"] = df["Order Date"].dt.to_period("M")

df = (
    df.groupby(["Customer Name", "OrderMonth"])["Sub Category"]
      .apply(lambda x: ",".join(sorted(set(x))))
      .reset_index(name="products")
)

# Keep only the column we need
df = df[["products"]]
df.head()
```

```python
df.shape #Shows the rows and columns
```

```python
# Turns into a list of items
data = list(df["products"].apply(lambda x: x.split(",")))
data
```

```python
# Replaces the true ands false and makes turns them into 1 or 0(one-hot encoding)
# Columns = items, rows = baskets

from mlxtend.preprocessing import TransactionEncoder
a = TransactionEncoder()
a_data = a.fit(data).transform(data)
df = pd.DataFrame(a_data, columns=a.columns_)

# Turn False/True into 0/1
df = df.replace(False, 0)
df = df.replace(True, 1)

df
```

```python
# Finds the items groups that show up often
df = apriori(df, min_support=0.05, use_colnames=True)
df
```

```python
# confidence rule.
df_ar = association_rules(df, metric="confidence", min_threshold=0.30)
df_ar
```

```python
# Lift rules
df_ar = association_rules(df, metric="lift", min_threshold=1)
df_ar
```

```python
# Conviction rules
df_ar = association_rules(df, metric="conviction", min_threshold=1.0)
df_ar
```

The Supermart dataset contains orders placed by customers on a grocery app. We grouped the data by customers and months, then converted the column into a list. We then transformed the list with one-hot encoding to represent each transaction. We viewed the calculations of support, confidence, lift, and conviction to view any purchase behaviors or patterns with each product/purchase. The most frequent item that was purchased was healthy drinks with a value of .27. The probability of organic staples and health drinks being purchased was at a high confidence level of 33 and implies that if a person buys organic staples there is a 33% chance that a healthy drink will be purchased. We can see a positive correlation with soft drinks and bread being purchased with a lift of 1.27 and are more likely to be bough together. By looking at the conviction rule, we can see that bread and soft drinks are positively associated, and both of these items directly impact each other with a conviction of 1.1.
