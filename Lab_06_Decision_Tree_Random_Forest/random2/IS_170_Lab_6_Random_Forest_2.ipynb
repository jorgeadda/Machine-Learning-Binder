```python
# Jorge Angon, Faith Yang
# Random Forest Model 2
# Dataset: https://www.kaggle.com/datasets/jayjoshi37/online-learning-engagement-and-dropout-risk

# import libraries
import pandas as pd
import kagglehub

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
from sklearn.preprocessing import LabelEncoder

# load dataset
path = kagglehub.dataset_download(
    "jayjoshi37/online-learning-engagement-and-dropout-risk"
)

df = pd.read_csv(path + "/online_learning_engagement_dropout_risk.csv")

print("Dataset Preview:")
print(df.head())

# feature engineering
# convert categorical variables to numbers
label_encoder = LabelEncoder()

for column in df.select_dtypes(include=["object"]).columns:
    df[column] = label_encoder.fit_transform(df[column])

# check for missing values and fill with column mean
df = df.fillna(df.mean())

# preprocessing: convert categorical features to numerical using one-hot encoding
df = pd.get_dummies(df, columns=['course_category', 'device_type'], drop_first=True)

# separate features and target
# target variable: dropout risk
X = df.drop("dropout_risk", axis=1)
y = df["dropout_risk"]

# train/split test
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

# train random forest model
rf_model = RandomForestClassifier(
    n_estimators=100,      # number of trees
    criterion="gini",      # split measurement
    max_features="sqrt",   # random subset of features
    bootstrap=True,
    oob_score=True,
    random_state=42
)

rf_model.fit(X_train, y_train)

# make predictions
y_pred = rf_model.predict(X_test)

# model evaluation
accuracy = accuracy_score(y_test, y_pred)

print("Random Forest Accuracy:", accuracy)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# F1 Score
f1 = f1_score(y_test, y_pred, average='weighted')

print("\nF1 Score:", f1)

# prepare a provision for a new data entry and its prediction of an outcome.

def predict_dropout():
    while True:
        try:
            print("\nEnter new student engagement data to predict dropout risk:")

            # Create dictionary for input values
            new_data = {}

            # Ask user for each feature used in the model
            for feature in X.columns:
                value = float(input(f"Enter value for {feature}: "))
                if value < 0:
                    print("Error: Please enter non-negative values.")
                    raise ValueError
                new_data[feature] = value

            # Convert to dataframe
            new_student = pd.DataFrame([new_data])

            # Predict
            prediction = rf_model.predict(new_student)[0]

            if prediction == 1:
                print("Prediction: High Dropout Risk")
            else:
                print("Prediction: Low Dropout Risk")

            break

        except ValueError:
            print("Error: Invalid input. Please enter numeric values only.")

# Run prediction
predict_dropout()

# feature importance
feature_importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf_model.feature_importances_
}).sort_values(by="Importance", ascending=False)

print("\nFeature Importance:")
print(feature_importance)

# example prediction
new_student = X_test.iloc[[0]]

prediction = rf_model.predict(new_student)

print("\nPredicted Dropout Risk:", prediction[0])

# This code uses a Random Forest model to predict whether a student is at risk of dropping out of an online course.
# The dataset is cleaned and prepared by converting text values to numbers and handling missing data.
# The data is then split into training and testing groups so the model can learn patterns and be evaluated.
# After training, the model makes predictions and its performance is measured using accuracy, a confusion matrix, and an F1 score.
# The program also lets a user enter new student information to predict if that student has a high or low dropout risk.
```
