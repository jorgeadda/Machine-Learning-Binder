```python
#CSV upload and how to load
import pandas as pd
import numpy as np

from google.colab import files
uploaded = files.upload() #Lets me choose a file

df=pd.read_csv('food_ingredients_and_allergens.csv') #reads the csv file into the pandas dataframe

df.head() #this displays the dataset
```

```python
#will save a copy of the original and find blanks, fill them and remove rows that are missing from prediction
df_original = df.copy()

df.isnull().sum() #Checks if there are missing values in each column

df1 = df.fillna('None') # fills missing value with a placeholder so the rows wont get lost
df = df1

df = df[df['Prediction'] != 'None'] #This removes rows if prediction is missing

df.shape
```

```python
#Turning the prediction text into numbers
from sklearn.preprocessing import LabelEncoder #Converts text to integers

le = LabelEncoder()

label = le.fit_transform(df['Prediction']) #Converts Prediction Column into a numeric labels

label #views the label
```

```python
#Replacing the text prediction column with numbered one

df.drop('Prediction', axis = 1, inplace = True)
df.drop('Allergens', axis = 1, inplace = True) #Drops allergens as well since it gives answer for prediction

df['Prediction'] = label #Adds prediction back as a numerical label instead

df.head()
```

```python
#Turn ingredients into their own columns (hot encoding)

!pip -q install mlxtend
from mlxtend.preprocessing import TransactionEncoder

#Changes these into a list
data = df[['Food Product', 'Main Ingredient', 'Sweetener', 'Fat/Oil', 'Seasoning']].values.tolist()

#Fits and turns into true or false
a = TransactionEncoder()
a_data = a.fit(data).transform(data)

df_hot = pd.DataFrame(a_data, columns=a.columns_) #Changes to a data frame with column names
df_hot = df_hot.replace(False, 0) # Makes false = 0
df_hot = df_hot.replace(True, 1) # Makes true = 1

df_hot.head()
```

```python
#Combining the new ingredient columns with a prediction
df2 = df_hot.copy() #Adds numeric target column to the hot style encoding
df2['Prediction'] = df['Prediction']

df2.head()
```

```python
#Scaling

from sklearn import preprocessing

x = df2.drop('Prediction', axis = 1) #Change to x to make it easier to type. Also drops the target

df_scaled = preprocessing.scale(x) #scales x

df_scaled = pd.DataFrame(df_scaled, columns = x.columns) #Converts scaled array back to a data frame with same column names

df_scaled['Prediction'] = df2['Prediction'] #Adds the target back

df.head()
```

```python
#Moving Food Product column to the end of the data set

OG_col = df.columns.values #orignal column
new_col = list(OG_col) #New column

new_col.remove('Food Product') #Removes food product

new_col.append('Food Product') #Adds food product at the end

df = df.reindex(columns = new_col) #Reorders the data frame columns

df.head()
```
