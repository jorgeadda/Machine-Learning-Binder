```python
import pandas as pd # importing both panda and numpy library
import numpy as np
```

```python
from google.colab import files #starting code to import files into google colab
uploaded=files.upload()#Gives you prompt to upload dataset
df=pd.read_csv('machine-readable-business-employment-data-sep-2025-quarter.csv')#must recognize correct file before uploading dataset
df.head(100)#displays the first 100 rows
```

```python
df.isnull().sum()#Shows the number of NA in each column
```

```python
df.fillna(value=0, inplace=True) #Fills in NA spots and replace it with 0
df.isnull().sum() #shows zero NA by replacing it with 0
```

```python
df.head(100)
```

```python
from sklearn.preprocessing import LabelEncoder #This is importing the LabelEncoder library from sklearn
le = LabelEncoder() #this creates or sets an intance for le for label encoder
label= le.fit_transform(df['STATUS'])#putting .fit_transformed fits label
label
```

```python
df.drop("STATUS", axis=1, inplace=True) #droping the STATUS column from df for now
df["STATUS"] = label #putting the STATUS back into the dataset ut attaching the array to it with the same name
df.head()
```

```python
#Same process but using for the Series_title_3 column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
label =le.fit_transform(df['Series_title_3'])
label
```

```python
#Same process but using the Series_title_3 column
df.drop("Series_title_3", axis=1, inplace=True)
df["Series_title_3"] = label
df.head(100)
```

```python
#This code renaranges the columns
# Originally, STATUS was at the very end so I moved it in front of the series
old_cols=df.columns.values
new_cols=['Series_reference', 'Period', 'Data_value', 'Suppressed', 'UNITS', 'Magnitude', 'Subject', 'Group', 'STATUS', 'Series_title_1', 'Series_title_2', 'Series_title_3', 'Series_title_4', 'Series_title_5']
df=df.reindex(columns=new_cols)
df
```
