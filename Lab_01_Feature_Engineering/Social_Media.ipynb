```python
import pandas as pd
import numpy as np
```

```python
#Uploading dataset to google colab
#Dataset includes data about students classes, academic level, etc. It shows how addicted they can be to their phones.
from google.colab import files
uploaded=files.upload()
df=pd.read_csv('Students Social Media Addiction.csv')
df.head(100)
```

```python
#There are no NA in this data set as everything shows 0 null columns
df.isnull().sum()
```

```python
#Using the Label Encoder to transform the affects academic performance because it fits the true/false category
#it displays the array. It will replace all the trues with 0s and false with 1s
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
label=le.fit_transform(df['Affects_Academic_Performance'])
label
```

```python
#It will drop the columns and basically put it back into table with the new labels
df.drop("Affects_Academic_Performance", axis=1, inplace=True)
df["Affects_Academic_Performance"]=label
df
```

```python
#We decided to do the same for Gender as there was only two genders listed
#This is the same process as before for affects adameci performance
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
label=le.fit_transform(df['Gender'])
label
```

```python
#Same process as it drops Genders and reenters the table with the label encoder
df.drop("Gender", axis=1, inplace=True)
df["Gender"]=label
df
```

```python
#Since we dropped the last two columns they reappear at the very end of the tabel and out of order
#We used this code reorder the table and place "Gender" and "Affects_Academic_Performance"
old_cols=df.columns.values
new_cols=['Student_ID', 'Age', 'Gender','Academic_Level', 'Country', 'Avg_Daily_Usage_Hours', 'Most_Used_Platform', 'Sleep_Hours_Per_Night', 'Mental_Health_Score', 'Relationship_Status', 'Conflicts_Over_Social_Media', 'Affects_Academic_Performance', 'Addicted_Score'  ]
df=df.reindex(columns=new_cols)
df
```

```python
#This just renames the columns names
#For example, we renamed Avg_Daily_Usage_Hours to Avg_Daily_Hours
print(df.rename(columns={'Avg_Daily_Usage_Hours':'Avg_Daily_Hours'}))
```
