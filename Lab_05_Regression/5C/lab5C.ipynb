```python
# Read the csv file
df = pd.read_csv('Medicaldataset.csv')
df
```

```python
# Check the data types
df.dtypes
```

```python
# Check the target values
df['Result'].value_counts()
```

```python
# Convert the Result column from words to numbers
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Result'] = le.fit_transform(df['Result'])
df
```

```python
# Split the dataset into x and y
# x = the 3 input columns
# y = the answer column
x = df[['Age', 'CK-MB', 'Troponin']].values
y = df['Result'].values
```

```python
# Split the data into training data and testing data
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=0
)
```

```python
# Scale the x data so the numbers are more balanced
from sklearn.preprocessing import StandardScaler

st_x = StandardScaler()
x_train = st_x.fit_transform(x_train)
x_test = st_x.transform(x_test)
```

```python
# Train the logistic regression model
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```

```python
# Make predictions using the test data
y_pred = classifier.predict(x_test)
y_pred
```

```python
# Check the model accuracy
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the model: {accuracy}")
```

```python
# Check precision, recall, and f1-score
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```

```python
# Let the user enter new patient data
while True:
    try:
        age = int(input("Enter Age: "))
        ck_mb = float(input("Enter CK-MB: "))
        troponin = float(input("Enter Troponin: "))

        new_data_point = nm.array([[age, ck_mb, troponin]])
        new_data_point = st_x.transform(new_data_point)

        prediction = classifier.predict(new_data_point)

        if prediction[0] == 0:
            print("Prediction: Negative")
        else:
            print("Prediction: Positive")

        break

    except ValueError:
        print("Invalid input. Please enter valid numeric values.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

For this dataset, we used binary logistic regression to predict whether a patient is at cardiac risk which is split into postive and negative. We used Age, CK-MB and Troponin becuase they are the most useful health measurements related to heart conditions. The model was trained using patient data to learn the patterns between the inputs and the final results. When tested, it had a 79% accuracy test whcih shows that that it usually is able to make the correct classification.
